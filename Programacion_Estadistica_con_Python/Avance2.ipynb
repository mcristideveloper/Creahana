```python
corrs = vino.corr()

mask = np.triu(np.ones_like(corrs, dtype=bool))
#sns.heatmap(corrs, vmin=-1, vmax=1, cmap="PiYG")
sns.heatmap(corrs, vmin=-1, vmax=1,
            cmap=sns.diverging_palette(110, 20, n=100, as_cmap=True),
            center=0, linewidths=0.0, mask=mask, linecolor=None)
```
